```python
# Task 2: Mark the success/failed launches for each site on the map
import folium
from folium import DivIcon

# Create a Folium map centered at NASA Johnson Space Center's coordinate
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Add a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
site_map.add_child(circle)

# Define marker icons for success and failure
success_icon = folium.Icon(icon='check', prefix='fa', color='green')
failure_icon = folium.Icon(icon='times', prefix='fa', color='red')

# Iterate through the launch sites and add markers for success and failure
for index, site in launch_sites_df.iterrows():
    site_name = site['Launch Site']
    site_coordinate = [site['Lat'], site['Long']]
    launch_outcome = site['class']
    
    # Determine the icon based on launch outcome
    if launch_outcome == 1:
        marker = folium.map.Marker(
            site_coordinate,
            icon=success_icon,
            popup=f'Successful Launch at {site_name}',
        )
    else:
        marker = folium.map.Marker(
            site_coordinate,
            icon=failure_icon,
            popup=f'Failed Launch at {site_name}',
        )
    
    site_map.add_child(marker)

# Save the map to an HTML file
site_map.save("launch_sites_map_with_outcomes.html")

# Display the map
site_map
```

```python
import folium

# Create a folium.Map
site_map = folium.Map(location=[28.563197, -80.57682], zoom_start=8)

# Group launch sites by site name
launch_sites = spacex_df['Launch Site'].unique()

# Create a dictionary to store marker clusters for each launch site
marker_clusters = {}

# Iterate through launch sites
for site in launch_sites:
    # Create a folium.MarkerCluster for each launch site
    marker_clusters[site] = MarkerCluster()

# Iterate through the rows of spacex_df
for index, record in spacex_df.iterrows():
    site_name = record['Launch Site']
    latitude = record['Lat']
    longitude = record['Long']
    launch_outcome = record['class']
    
    # Determine the marker color based on launch outcome
    marker_color = 'green' if launch_outcome == 1 else 'red'
    
    # Create a folium.Marker for the current record
    marker = folium.Marker(
        location=[latitude, longitude],
        icon=folium.Icon(color='white', icon_color=marker_color),
        popup=f'Launch Outcome: {"Success" if launch_outcome == 1 else "Failure"}'
    )
    
    # Add the marker to the corresponding marker cluster based on launch site
    marker_clusters[site_name].add_child(marker)

# Add the marker clusters to the site_map for each launch site
for site_name, marker_cluster in marker_clusters.items():
    site_map.add_child(marker_cluster)

# Display the updated site_map with markers grouped by launch site
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance between two coordinates using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Approximate radius of Earth in kilometers
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Coordinates of the launch site (replace with actual launch site coordinates)
launch_site_lat = 28.563197
launch_site_lon = -80.57682

# Coordinates of the closest point of interest (replace with actual coordinates)
poi_lat = 28.56367
poi_lon = -80.57163

# Calculate the distance between the launch site and the point of interest
distance_poi = calculate_distance(launch_site_lat, launch_site_lon, poi_lat, poi_lon)

# Create a marker for the point of interest with distance popup
poi_marker = folium.Marker(
    location=[poi_lat, poi_lon],
    icon=folium.DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12; color:#d35400;"><b>{distance_poi:.2f} KM</b></div>'
    ),
    popup=f'Closest Point of Interest (Lat: {poi_lat}, Lon: {poi_lon})'
)

# Create a PolyLine between the launch site and the point of interest
line_coordinates = [(launch_site_lat, launch_site_lon), (poi_lat, poi_lon)]
poi_polyline = folium.PolyLine(
    locations=line_coordinates,
    weight=1,  # Set the line weight (thickness)
    color='blue'  # Set the color of the PolyLine
)

# Add the marker and PolyLine to the site_map
site_map.add_child(poi_marker)
site_map.add_child(poi_polyline)

# Display the updated site_map with the marker and PolyLine
site_map
```
